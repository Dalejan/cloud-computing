```python
import pandas as pd

fullCorpus = pd.read_csv("SMSSpamCollection.tsv", sep="\t", header=None)
fullCorpus.columns= ['label', 'body_text']

fullCorpus.head()
```

```python
print(f"Input data has {len(fullCorpus)} rows and {len(fullCorpus.columns)} columns")

print(f"\nThere are {len(fullCorpus[fullCorpus['label']=='spam'])} samples are spam")
print(f"\nThere are {len(fullCorpus[fullCorpus['label']=='ham'])} samples are ham")

```

```python
print(f"\nThere are {(fullCorpus['label']).isnull().sum()} label loses")
print(f"\nThere are {(fullCorpus['body_text']).isnull().sum()} body_text loses")

```

### Exploring the data with facets

```python

from IPython.core.display import display, HTML
import tensorflow as tf
import base64
from facets_overview.feature_statistics_generator import FeatureStatisticsGenerator

fsg = FeatureStatisticsGenerator()
dataframes = [
    {'table': fullCorpus, 'name': 'Dataset'}]
censusProto = fsg.ProtoFromDataFrames(dataframes)
protostr = base64.b64encode(censusProto.SerializeToString()).decode("utf-8")


HTML_TEMPLATE = """<script src="https://cdnjs.cloudflare.com/ajax/libs/webcomponentsjs/1.3.3/webcomponents-lite.js"></script>
        <link rel="import" href="https://raw.githubusercontent.com/PAIR-code/facets/1.0.0/facets-dist/facets-jupyter.html">
        <facets-overview id="elem"></facets-overview>
        <script>
          document.querySelector("#elem").protoInput = "{protostr}";
        </script>"""
html = HTML_TEMPLATE.format(protostr=protostr)
display(HTML(html))
```
